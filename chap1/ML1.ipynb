### 9. What is an online learning system?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
In online learning, you train the system incrementally by feeding it data instances sequentially, either individually or in small groups called mini-batches. Each learning  step is fast and cheap, so the system can learn about new data on the fly, as it arrives.
</p>

### 10. What is out-of-core learning?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Online learning algorithms that can also be used to train systems on huge datasets that cannot fit in one machine’s main memory. The algorithm loads part of the data, runs a training step on that data, and repeats the  process until it has run on all of the data.
</p>

### 12. What is the difference between a model parameter and a learning algorithm’s hyperparameter?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
A model parameter is a parameter of the function/model that is calculated while traingin to fit the data, whereas a hyperparameter is a parameter of a learning algorithm, it is not affected by the algorithm calculations itself since it is set prior to training and remain constant.
</p>

### 15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?
<p>
The model is overffiting the data, it is too complex relative to the amount and noisiness of the training data. For regularization, possible solutions are:
</p>
<p>
• Simplify the model by selecting one with fewer parameters (e.g., a linear model rather than a high-degree polynomial model), by reducing the number of attributes in the training data, or by constraining the model.
</p><p>
• Gather more training data.
</p><p>
• Reduce the noise in the training data (e.g., fix data errors and remove outliers).
</p>

### 16. What is a test set, and why would you want to use it?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Test set is the data on which the trained model is tested or evaluated. By evaluating the model on test set we get an estimate of the error rate on new cases also known as generalization error.
</p>

### 17. What is the purpose of a validation set?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
The purpose of a validation set is to hold out a part of the training set to evaluate several candidate models and select the best model in the end.
</p>

### 18. What is the train-dev set, when do you need it, and how do you use it?
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
The train-dev set is used when there is a risk of mismatch between the training data and the data used in the  validation and test datasets (which should always be as close as possible to the data used once the model is in production). The train-dev set is a part of the training set that’s held out (the model is not trained on it). The model is trained on the rest of the training set, and evaluated on both the train-dev set and the validation set. If the model performs well on the training set but not on the train-dev set, then the model is likely overfitting the training set. If it performs well on both the training set and the train-dev set, but not on the validation set, then there is probably a significant data mismatch between the training data and the validation + best data, and you should try to improve the training data to make it look more like the validation + test data.
</p>

### 19. What can go wrong if you tune hyperparameters using the test set? 
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Overfitting the test set. The generalization error will be optimistic.
</p>
